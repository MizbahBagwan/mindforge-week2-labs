```python
import pandas as pd
import numpy as np

data = {
    "Intersection": ["Navi Peth", "Saat Rasta", "Mechanic Chowk", "Saat Rasta", "Navi Peth", "Mechanic Chowk"],
    "Time": ["08:00 AM", "08:00 AM", "08:00 AM", "09:00 AM", "09:00 AM", "09:00 AM"],
    "Vehicle_Count": [450, -20, 310, np.nan, 600, 420],
    "Sensor_Status": ["Active", "Faulty", "Active", "Offline", "Active", "Active"]
}

df = pd.DataFrame(data)

print(df)
```

```python
df.loc[df["Vehicle_Count"] < 0, "Vehicle_Count"] = 0

print(df)
```

```python
median_value = df["Vehicle_Count"].median()

df["Vehicle_Count"] = df["Vehicle_Count"].fillna(median_value)

print(df)
```

```python
# Total vehicles by intersection
traffic_summary = df.groupby("Intersection")["Vehicle_Count"].sum()

print(traffic_summary)
```

```python
busiest = traffic_summary.idxmax()

print("The busiest intersection is:", busiest)
print("It requires immediate AI optimization.")
```
