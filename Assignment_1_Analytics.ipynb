```python
import pandas as pd
import numpy as np

# Simulating 50,000 retail transactions
np.random.seed(42)

data = {
    "Transaction_ID": range(10000, 60000),
    "Store_City": np.random.choice(["Mumbai", "Pune", "Delhi", "Bangalore"], 50000),
    "Category": np.random.choice(["Electronics", "Clothing", "Groceries"], 50000),
    "Revenue": np.random.normal(500, 150, 50000)
}

df = pd.DataFrame(data)

# Injecting corruption (missing values)
df.loc[df.sample(frac=0.05).index, "Revenue"] = np.nan

df.head()
```

```python
print("Missing Revenue values:", df["Revenue"].isna().sum())
```

```python
# Calculate median revenue
median_revenue = df["Revenue"].median()

print("Median Revenue:", median_revenue)

# Fill missing values with median
df["Revenue"] = df["Revenue"].fillna(median_revenue)

# Verify missing values are gone
print("Missing Revenue values after cleaning:", df["Revenue"].isna().sum())
```

```python
city_revenue = df.groupby("Store_City")["Revenue"].sum()

print(city_revenue)
```

```python
highest_city = city_revenue.idxmax()

print("City with highest total revenue:", highest_city)
```
