```python
server_logs = """
[INFO] 192.168.1.1 - User Logged In
[ERROR] 10.0.0.5 - Database Connection Failed
[INFO] 192.168.1.2 - Page Loaded
[CRITICAL] 172.16.0.4 - Server Overheating!
[ERROR] 10.0.0.9 - Password Invalid
"""

print(server_logs)
```

```python
lines = server_logs.strip().split("\n")

print(lines)
```

```python
critical_issues = []
```

```python
for line in lines:

    if "[ERROR]" in line or "[CRITICAL]" in line:

        parts = line.split(" - ")

        left_part = parts[0]
        message = parts[1]

        ip_address = left_part.split()[1]

        critical_issues.append((ip_address, message))
```
